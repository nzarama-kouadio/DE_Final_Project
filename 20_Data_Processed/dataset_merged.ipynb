```python
# Step 2: Visualize all of them

# Loop through the dictionary and print the head of each DataFrame
for name, df in dataframes.items():
    print(f"Head of {name}:")
    print(df.head())
    print("\n" + "-" * 50 + "\n")
```

```python
# Step 4: Merge account_activity_data and customer_data, suspicious_activity_data on "CustomerID"

# List of DataFrames to merge
dataframes = [customer_data, account_activity_data, suspicious_activity_data]

# Start with the first DataFrame
customer_merge_data = dataframes[0]

# Loop through the remaining DataFrames and merge them sequentially
for df in dataframes[1:]:
    customer_merge_data = pd.merge(
        customer_merge_data, df, on="CustomerID", how="inner", validate="one_to_one"
    )

# Display the merged DataFrame
print("Merged DataFrame using loop:")
customer_merge_data.head()
```

```python
# Step 5: Merge datasets involving transaction Data: transaction_meta_data, transaction_records_data, amount_data, anomaly_scores_data, fraud_indicators_data, transaction_category_data

dataframes_2 = [
    transaction_meta_data,
    transaction_records_data,
    amount_data,
    anomaly_scores_data,
    fraud_indicators_data,
    transaction_category_data,
]

transaction_merge_data = dataframes_2[0]

# Loop through the remaining DataFrames and merge them sequentially
for df in dataframes_2[1:]:
    transaction_merge_data = pd.merge(
        transaction_merge_data,
        df,
        on="TransactionID",
        how="inner",
        validate="one_to_one",
    )

# Display the merged DataFrame
print("Merged DataFrame using loop:")
transaction_merge_data.head()
```

```python
# Step 6: Merge transaction_merge_data and merchant_data based on MerchantID, use many_to_one

transaction_merge_data = pd.merge(
    transaction_merge_data,
    merchant_data,
    on="MerchantID",
    how="inner",
    validate="many_to_one",
)
transaction_merge_data.head()
```

```python
# Step 8: Finally merge transaction_merge_data and customer_merge_data based on "CustomerID" use many_to_one
complete_dataset = pd.merge(
    transaction_merge_data,
    customer_merge_data,
    on="CustomerID",
    how="inner",
    validate="many_to_one",
)
complete_dataset.head()
```

```python
# Rename columns
complete_dataset.rename(
    columns={
        "Name": "CustomerName",
        "Location": "MerchantLocation",
        "Age": "CustomerAge",
        "Address": "CustomerAddress",
    },
    inplace=True,
)
complete_dataset.head()
```

```python
# Sort dataset by transaction to make it look more intuitive
complete_dataset = complete_dataset.sort_values(by="TransactionID")
complete_dataset.head()
```

```python
# Make sure that transaction_ID is the true unique index
# Find the non-unique TransactionID values
non_unique_transaction_id = complete_dataset["TransactionID"][
    complete_dataset["TransactionID"].duplicated(keep=False)
]

print(non_unique_transaction_id.unique())
```

```python
# Save your complete dataset to a CSV file
complete_dataset.to_csv("complete_dataset.csv", index=False)
print("File created successfully? :", os.path.exists("complete_dataset.csv"))
```
